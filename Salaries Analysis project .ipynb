```python
plt.figure(figsize=(10,6))
sns.countplot(x='company_location',data=df, hue='work_year')
plt.xticks(rotation=90)
plt.tight_layout()
plt.title('location hired most')
```

```python
# Most hiring companies by scale size
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x='company_size',data=df, hue='employment_type')
plt.xticks(rotation=90)
plt.legend(loc='upper right',bbox_to_anchor=(1.1,1.0))
plt.tight_layout()
plt.title('Most Hired Job type')
```

```python
plt.figure(figsize=(11,9))
sns.countplot(x='job_title',data=df, hue='work_year')
plt.xticks(rotation=90)
plt.tight_layout()
plt.title('Most Hired Job title')
```
